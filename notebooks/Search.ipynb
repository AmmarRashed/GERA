```python
from functools import reduce

temp = [('(', ' '), (')', ' '), ('-', ' ')]
cancers = []

for k in search.split('\n\n')[::2]:
    k = reduce(lambda z, rp: z.replace(*rp), temp, k)
    if 'Select' not in k:
        cancers.append(k)
    
    else:
        cancers.append(k.split('\n')[-1])
```

```python
rest = set(cancers)
for d, c in product(done, cancers):
    if set(d.lower().split()) & set(c.lower().split()):
        rest -= set([c])
```
