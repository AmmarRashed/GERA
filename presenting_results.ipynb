```python
import pickle, pandas as pd
```

```python
results = pickle.load(open("results2.pickl", "rb"))
labels = pickle.load(open("datasets/gene_data_y","rb"))
```

```python
df_raw = dict()
for i in range(len(results)):
    for k in results[i]:
        df_raw.setdefault(k, [])
        df_raw[k].append(results[i][k])
```

```python
df = pd.DataFrame(df_raw).T
df.columns = labels
df
```

```python
df.to_csv("toy_data_results.csv", index_label="Pathway")
```
